# Python版くまじゃんけん

Python版のくまじゃんけんです。できるだけScratch版 (https://scratch.mit.edu/projects/344795135) と同じアルゴリズムになるように作りました。また、ちょっと難しいかな？と思う箇所にはコメントを追加してありますので、ご参考になさってください。

```
# くまじゃんけん
import random   # 乱数（ランダム）を扱うためのライブラリをインポート

kuma = 0    # クマの手を覚える変数
jibun = 0     # 自分の手を覚える変数
count = 10  # 何回勝負か決めておく
kumalist = []   # クマの手の履歴を残す
jibunlist = []    # 自分の手の履歴を残す

# janken関数
# 引数 te に数字(1 or 2 or 3)が渡ってくると message に
# ぐー, ちょき, ぱーのいずれかを入れて戻値（返値）で返します。
def janken(te):
  message = ''
  if te == 1:
    message = 'ぐー'
  elif te == 2:
    message = 'ちょき'
  else:
    message = 'ぱー'
  return message
```

```
# shoubu関数
# 引数 kakegoe に、じゃんけんのときの掛け声が渡ってきます。
# 戻値（返値）はありません。
def shoubu(kakegoe):
  global kuma
  global jibun
  print(kakegoe)
  kuma = random.choice([1, 2, 3])   # リスト[1, 2, 3] のうち一つをランダムに選びます
  jibun = ''
  # 変数 jibun に3より大きい数字や、数字が入ってなかったら input を繰り返します
  while not jibun.isdigit() or int(jibun) > 3:
    jibun = input('ぐー: 1, ちょき: 2, ぱー: 3  ')
  jibun = int(jibun)    # input の返値は文字列なので、数値に変換します
  print('〜〜〜〜〜〜〜〜')
  kumanote = janken(kuma)
  jibunnote = janken(jibun)
  print('くま: ' + kumanote, 'じぶん: ' + jibunnote)
  print('')
  kumalist.append(kuma)
  jibunlist.append(jibun)
```

```
# じゃんけんプログラムを実行します。
print('じゃんけん', count, '回勝負だ！')

for i in range(count):

  shoubu('じゃんけん')

  # あいこチェック
  hikaku = kuma - jibun
  while hikaku == 0:
    shoubu('あいこで')
    hikaku = kuma - jibun
    if hikaku != 0:
      break

  # 勝敗チェック
  if hikaku == 1 or hikaku == -2:
    print('くま：まけた〜')
  else:
    print('くま: かったー！')
  print('')
  print('')

print(kumalist)
print(jibunlist)

```

```
# じゃんけんの手の履歴をCSVファイルでダウンロードします。
import csv

# kuma.csv というファイル名で、書き込みモードのファイルを開きます
# csv.writer を使って、kumafile リストの中身をカンマ区切りで書き込みます
filename = 'kuma.csv'
with open(filename, 'w', newline='') as kumafile:
     wr = csv.writer(kumafile, quoting=csv.QUOTE_ALL)
     wr.writerow(kumalist)
  
# jibun.csv というファイル名で、書き込みモードのファイルを開きます
# csv.writer を使って、jibunfile リストの中身をカンマ区切りで書き込みます
filename = 'jibun.csv'
with open(filename, 'w', newline='') as jibunfile:
     wr = csv.writer(jibunfile, quoting=csv.QUOTE_ALL)
     wr.writerow(jibunlist)

# google.colab のライブラリを使ってファイルをローカルに保存します。
from google.colab import files

files.download('kuma.csv')
files.download('jibun.csv')
```
